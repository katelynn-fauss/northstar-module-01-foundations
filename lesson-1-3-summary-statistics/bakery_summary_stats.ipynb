```python
import pandas as pd
import numpy as np
```

```python
bakery = pd.read_csv("../lesson-1-2-types-of-data/bakery_customers.csv")
bakery.head(10)
```

```python
total_spent_mean = bakery["total_spent_usd"].mean()
print(f"Mean total spent: ${total_spent_mean:.2f}")
```

```python
total_spent_median = bakery["total_spent_usd"].median()
print(f"Median total spent: ${total_spent_median:.2f}")
```

```python
total_spent_mode = bakery["total_spent_usd"].mode()
print(f"Mode total spent: {total_spent_mode.values}")
```

```python
satisfaction_mode = bakery["satisfaction"].mode()
print(f"Mode satisfaction: {satisfaction_mode.values}")
```

```python
satisfaction_median = bakery["satisfaction"].median()
print(f"Median satisfaction: {satisfaction_median}")
```

```python
satisfaction_mean = bakery["satisfaction"].mean()
print(f"Mean satisfaction: {satisfaction_mean:.2f}")
print("(Mean is technically a category error on ordinal data; use with care.)")
```

```python
region_mode = bakery["region"].mode()
print(f"Most common region: {region_mode.values[0]}")
```

```python
spent_min = bakery["total_spent_usd"].min()
spent_max = bakery["total_spent_usd"].max()
spent_range = spent_max - spent_min
spent_variance = bakery["total_spent_usd"].var()
spent_std = bakery["total_spent_usd"].std()
print(f"Min: ${spent_min:.2f}")
print(f"Max: ${spent_max:.2f}")
print(f"Range: ${spent_range:.2f}")
print(f"Variance: ${spent_variance:.2f} (squared dollars)")
print(f"Standard deviation: ${spent_std:.2f}")
```

```python
spent_q1 = bakery["total_spent_usd"].quantile(0.25)
spent_q3 = bakery["total_spent_usd"].quantile(0.75)
spent_iqr = spent_q3 - spent_q1
print(f"Q1 (25th percentile): ${spent_q1:.2f}")
print(f"Q3 (75th percentile): ${spent_q3:.2f}")
print(f"IQR: ${spent_iqr:.2f}")
```

```python
spent_p90 = bakery["total_spent_usd"].quantile(0.90)
spent_p95 = bakery["total_spent_usd"].quantile(0.95)
spent_p99 = bakery["total_spent_usd"].quantile(0.99)
print(f"90th percentile: ${spent_p90:.2f}")
print(f"95th percentile: ${spent_p95:.2f}")
print(f"99th percentile: ${spent_p99:.2f}")
```
